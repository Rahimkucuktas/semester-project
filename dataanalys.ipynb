```python
#8. What are the 3 most popular databases ?¶

values = list()
for row in df['DatabaseWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values.append(value)
            
df1 = pd.DataFrame(values,columns=['Databases'])
df2=df1['Databases'].value_counts()
df2[0:3]
```

```python
#9. Is there a difference between programming languages when it comes to job satisfaction?¶


df1=df[['JobSat','LanguageWorkedWith']]



filtered_df = df1[df1['JobSat'] == 'Very satisfied'].dropna()


values = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values.append(value)
values

df20 = pd.DataFrame(values,columns=['Languages'])
df20.value_counts()
#Very Satisfied programmers uses this languages...
```

```python

#very satisfied 
df1=df[['JobSat','LanguageWorkedWith']]

filtered_df = df1[df1['JobSat'] == 'Very satisfied'].dropna()

values1 = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values1.append(value)
        
values1

df10 = pd.DataFrame(values1,columns=['Languages'])
point5=df10.value_counts()*5

#Slightly satisfied 


df2=df[['JobSat','LanguageWorkedWith']]

filtered_df = df2[df2['JobSat'] == 'Slightly satisfied'].dropna()


values2 = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values2.append(value)
            
values2

df20 = pd.DataFrame(values2,columns=['Languages'])
point4=df20.value_counts()*4

#Neither satisfied nor dissatisfied

df3=df[['JobSat','LanguageWorkedWith']]

filtered_df = df3[df3['JobSat'] == 'Neither satisfied nor dissatisfied'].dropna()


values3 = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values3.append(value)
values3

df30 = pd.DataFrame(values3,columns=['Languages'])
point3=df30.value_counts()*3

#Slightly dissatisfied

df4=df[['JobSat','LanguageWorkedWith']]

filtered_df = df4[df4['JobSat'] == 'Slightly dissatisfied'].dropna()

values4 = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values4.append(value)
values4

df40 = pd.DataFrame(values4,columns=['Languages'])
point2=df40.value_counts()*2

#very dissatisfied

df5=df[['JobSat','LanguageWorkedWith']]

filtered_df = df5[df5['JobSat'] == 'Very dissatisfied'].dropna()

values5 = list()
for row in filtered_df['LanguageWorkedWith']:
    if not pd.isna(row):
        for value in row.split(";"):
            values5.append(value)
values5

df50 = pd.DataFrame(values5,columns=['Languages'])
point1=df50.value_counts()

#total point of the languages
total_point=(point1+point2+point3+point4+point5)
total_point


#this function can calculate each language point
def languageSat(x):
    d=total_point[x]/[values1.count(x)+values2.count(x)+values3.count(x)+values4.count(x)+values5.count(x)]
    print(d)
    
languageSat('Java')
```
